```python

df = pd.DataFrame({
    'Order Date': ['2023-01-10', '2023-03-15', '2023-07-25']
})
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month_name()
df['Year'] = df['Order Date'].dt.year

print(df)
```

```python


df = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Furniture', 'Electronics'],
    'Sales': [500, 300, 400, 700]
})

sns.barplot(x='Category', y='Sales', data=df)
plt.title('Sales by Category')
plt.show()
```

```python
df = pd.DataFrame({
    'Region': ['East', 'West', 'North', 'South', 'East', 'West'],
    'Month': ['Jan', 'Feb', 'Mar', 'Jan', 'Feb', 'Mar'],
    'Segment': ['Consumer', 'Corporate', 'Home Office', 'Consumer', 'Corporate', 'Home Office'],
    'Sales': [200, 300, 250, 400, 350, 450]
})
```

```python

df = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr'],
    'Sales': [300, 400, 500, 600]
})

df.plot(x='Month', y='Sales', kind='line')
plt.title('Monthly Sales Trend')
plt.show()

# Observation:
# Sales increase from Jan to Apr.
```

```python


df = pd.DataFrame({'Segment':['Consumer','Corporate','Home Office'],
                   'Sales':[400,500,700]})

sns.barplot(x='Segment', y='Sales', data=df)
plt.title('Sales by Segment')
plt.show()

# Observation:
# Home Office has highest sales.
```
